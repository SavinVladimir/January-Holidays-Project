```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize' : (9, 6)})
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
data = data.rename(columns = {'Annual Income (k$)':'Annual_Income', 'Spending Score (1-100)':'Spending_Score'})
```

```python
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.groupby('Gender').Age.agg(['mean', 'max', 'min'])
```

```python
data.groupby('Gender').Annual_Income.agg(['mean', 'max', 'min', 'var'])
```

```python
data.groupby('Gender').Spending_Score.agg(['mean', 'max', 'min', 'var'])
```

```python
size = data['Gender'].value_counts()

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors = ['lightgreen', 'orange'], 
            explode = [0, 0.1], 
            labels = ['Female', 'Male'], 
            shadow = True, 
            autopct = '%.2f%%')

plt.title('Gender', fontsize = 20)
plt.axis('off')
plt.legend()
```
